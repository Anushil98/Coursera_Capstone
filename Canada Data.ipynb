```python
#getting the required libraries
import requests
import bs4
```

```python
res=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```python
type(soup)
```

```python
table=soup.select('table')
```

```python
type(table)
```

```python
h=table[0]
```

```python
df=h.text.split("\n")
```

```python
len(df)
```

```python
df.remove("")#removing the empty strings
```

```python
 df=list(filter(lambda a: a != "", df))#removing all the empty strings
```

```python
df
```

```python
len(df)
```

```python
df_new=[]
i=0
for j in range(len(df)//3):
    df_new.append(df[i:i+3])
    i=i+3
```

```python
import pandas as pd
df_new=pd.DataFrame(df_new)
df_new
```

```python
column=df_new.loc[0][:]
type(column)
```

```python
df_new.columns=column
```

```python
df_new=df_new.drop(0,axis=0)
```

```python
df_new
```

```python
df_new=df_new.reset_index()
```

```python
df_new=df_new.drop("index",axis=1)
```

```python
#removing the rows with borough as not assigned
for index,i in enumerate(df_new.loc[:]["Borough"]):
    if i=="Not assigned":
        df_new=df_new.drop(index,axis=0)

```

```python
df_new=df_new.reset_index()
```

```python
df_new=df_new.drop("index",axis=1)
```

```python
df_new
```

```python
#grouping the dataFrame according to the PostalCode so that The Neighbouhood Of same Code Can be Combined
pc=df_new.groupby("Postcode")
pc
```

```python
df2=[]
for group in pc:
    h=group[1]
    n=list(h["Neighbourhood"])
    n=",".join(n[:])
    df2.append([h.iloc[0][0],h.iloc[0][1],n])

df2=pd.DataFrame(df2)
df2.columns=["Postcode","Borough","Neighbourhood"]
    
    
```

```python
df2
```

```python
#for neighbourhoods as not assigned the values ae changes to it's respective borough name
for index,i in enumerate(df2.loc[:]["Neighbourhood"]):
    if i=="Not assigned":
        df2.loc[index]["Neighbourhood"]=df2.loc[index]["Borough"]
    
```
